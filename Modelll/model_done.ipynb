```python
##Model stage 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
import pickle
```

```python
df_ = pd.read_csv('FinalData_edited.csv')
```

```python
df = df_.copy()
```

```python
df.sample(50)
```

```python
db = df.tail(50)
df.drop(df.tail(50).index,inplace=True)
df
```

```python
db.to_csv('DB_Data')
```

```python
df['parent_full_seq'] = df['parent_full_seq'].astype('category')
df['child_full_seq'] = df['child_full_seq'].astype('category')
```

```python
df_pre = df[['parent_full_seq','child_full_seq','target']]
```

```python
from sklearn.model_selection import train_test_split
X = df_pre.drop('target', axis=1) # input features
y = df_pre['target']              # target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Define the TfidfVectorizer with min_df=1 to include all k-mers
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(3,3), min_df=1)


# Generate the k-mer counts and transform them into embeddings


X_ch = vectorizer.fit_transform(X_train['child_full_seq'])
X_p = vectorizer.fit_transform(X_train['parent_full_seq'])


kmer_embeddings_c = X_ch.toarray()
kmer_embeddings_p = X_p.toarray()
```

```python
for i in range (0,len(kmer_embeddings_c[0])):
    X_train['child_gene_k_' + str(i)] = [kmer_embeddings_c[j][i] for j in range(0,len(X_train))]
    X_train['parent_gene_k_' + str(i)] = [kmer_embeddings_p[j][i] for j in range(0,len(X_train))]
```

```python
X_train.drop(columns=['parent_full_seq','child_full_seq'],inplace=True)
```

```python

import xgboost as xgb

# from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python


# Train XGBoost model
model_xgb = xgb.XGBClassifier(
 learning_rate=0.3,
    max_depth=15,
    n_estimators=15000,
    
)
model_xgb.fit(X_train, y_train)
```

```python
X_t_c = vectorizer.transform(X_test['child_full_seq'])
X_t_p = vectorizer.transform(X_test['parent_full_seq'])

kmer_embeddings_t_c = X_t_c.toarray()
kmer_embeddings_t_p = X_t_p.toarray()
```

```python
for i in range (0,len(kmer_embeddings_t_c[0])):
    X_test['child_gene_k_' + str(i)] = [kmer_embeddings_t_c[j][i] for j in range(0,len(X_test))]
    X_test['parent_gene_k_' + str(i)] = [kmer_embeddings_t_p[j][i] for j in range(0,len(X_test))]
```

```python
X_test.drop(columns=['parent_full_seq','child_full_seq'],inplace=True)
```

```python

y_pred = model_xgb.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 score:", f1_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```

```python

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


svm = SVC(kernel='rbf', C=0.5, gamma='scale') # use the radial basis function kernel with default hyperparameters

# Step 6: Train the SVM model on the training set
svm.fit(X_train, y_train)

# Step 7: Make predictions on the testing set
y_pred_svm = svm.predict(X_test)

# Step 8: Evaluate the performance of the model
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Precision:", precision_score(y_test, y_pred_svm))
print("Recall:", recall_score(y_test, y_pred_svm))
print("F1 score:", f1_score(y_test, y_pred_svm))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_svm))
```

```python
def compare_sequences(father, child):
    match_count = 0
    total_count = 0
    length = int(len(father) /2 )
    allel1_father = father[0:length]
    allel2_father = father[length+1 :]
    allel1_child = child[:length]
    allel2_child = child[length+1 :]
    # Compare allel1 sequences
    for i in range(len(allel1_father)):
        if allel1_father[i] == allel1_child[i]:
            match_count += 1
        total_count += 1

    # Compare allel2 sequences
    for i in range(len(allel2_father)):
        if allel2_father[i] == allel2_child[i]:
            match_count += 1
        total_count += 1

    # Calculate likelihood of paternity
    likelihood = (match_count / total_count) * 100

    return int(likelihood)

#Example usage
# allel1_father = df_e.iloc[6224,8]
# allel2_father = df_e.iloc[6224,9]

# allel1_child = df_G2001[df_G2001['ParentM']=='E6224'].iloc[1,8]
# allel2_child = df_G2001[df_G2001['ParentM']=='E6224'].iloc[1,9]

father = df_tt.iloc[42,1]
child = df_tt.iloc[42,2]

likelihood = compare_sequences(father , child)
print("Likelihood of paternity: {:.2f}%".format(likelihood))
```
